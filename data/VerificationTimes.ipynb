## Verification Times per Chain

**Chain verification times are grouped together**

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfVerTimes = pd.read_csv("RBV_verificationTimes.csv", delimiter=",", header=0)
# dfVerTimes
```

```python
# dfVerTimes.boxplot()
ax = sns.boxplot(data=dfVerTimes[~np.isnan(dfVerTimes)])

#Remove y axis
ax.set(xlabel=None)
ax.set(ylabel=None)

ax.yaxis.set_tick_params(labelsize = 14)
ax.xaxis.set_tick_params(labelsize = 14)

ax.set_xticklabels(ax.get_xticklabels(), rotation=20)

plt.savefig("verificationTimes.pdf",dpi=300)
```
